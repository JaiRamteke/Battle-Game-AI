```python
# @title
import random
import time
import matplotlib.pyplot as plt
class Character:
  def __init__(self, name, hp):
    self.name = name #Player 1, Player 2
    self.hp = hp # 100 HP
    self.is_defending = False

  def attack(self, target):
    damage = random.randint(10,20)
    if target.is_defending:
      damage //= 2
      target.is_defending = False
    target.hp = target.hp-damage

    print(f"{self.name} attacks {target.name} for {damage} damage!‚öîÔ∏è")
  def defend(self):
    self.is_defending = True
    print(f"{self.name} is defending!üõ°Ô∏è")

  def is_alive(self):
    if self.hp>0:
        return True
    else:
        return False


def computer_choice(computer, player):
    if player.is_defending:
        return "attack"

    if computer.is_defending:
        return "attack"

    if player.hp<30:
        return "attack"
    if computer.hp<20:
        return "defend"


    if random.random()<0.7: #70% chance of attacking
      return "attack"




def game():
    player1 = Character("Player 1", 100)
    player2 = Character("Player 2", 100)
    current_player = player1
    opponent = player2

    while player1.is_alive() and player2.is_alive():
            print("\n--- Turn Start ---")
            print(f"{player1.name} HP: {player1.hp}")
            print(f"{player2.name} HP: {player2.hp}")
            print(f"{current_player.name}'s turn:")

            while True:
                if current_player == player1:
                  action = input("Choose an action (attack, defend): ")
                else:
                  action = computer_choice(current_player,opponent)
                time.sleep(2)
                if action == "attack":
                    current_player.attack(opponent)
                    break
                elif action == "defend":
                    if current_player.is_defending:
                        print("Defenses are up! Choose another action.")
                        continue
                    current_player.defend()
                    break
                else:
                    print("Invalid action. Please choose again.")
                    continue
            current_player, opponent = opponent, current_player

    if player1.is_alive():
        print(f"{player1.name} wins!üèÜ")
    else:
        print(f"{player2.name} wins!")

game()
```
